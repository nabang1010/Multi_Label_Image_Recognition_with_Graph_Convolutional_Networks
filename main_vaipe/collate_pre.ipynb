```python
# range square img size 

img_list = [torch.randint(0, 255,(i, 20)) for i in range(10,17)]
img_label = torch.randint(0, 10, (7, ))
batch = list(zip(img_list, img_label))
def collate_fn_xa(batch):
# padding to max size
    batch_sample, batch_label = zip(*batch)
    max_len = max([i.shape[0] for i in batch_sample])
    n_ftrs = batch_sample[0].size(1)
    features = torch.zeros((len(batch_sample), max_len, n_ftrs))
    labels = torch.tensor(batch_label)
    for i in range(len(batch_sample)):
        j, k = batch_sample[i].size(0), batch_sample[i].size(1)
        features[i] = torch.cat([batch_sample[i], torch.zeros((max_len - j, k))])
    return features, labels

batch_out = collate_fn_xa(batch)
for img in batch_out[0]:
    print(img)
```
