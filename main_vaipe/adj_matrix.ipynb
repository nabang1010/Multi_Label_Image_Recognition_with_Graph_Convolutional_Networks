```python
# create a list of list of pill_name in description
pill_all = []
# /workspace/nabang1010/LBA_VAIPE/GNN/Multi_Label_Image_Recognition_with_Graph_Convolutional_Networks/data/public_train/pill/label/VAIPE_P_1170_21.json
for i in range(1173):
    pill_name_set = set()
    for j in range(150):
        json_path = "/workspace/nabang1010/LBA_VAIPE/GNN/Multi_Label_Image_Recognition_with_Graph_Convolutional_Networks/data/public_train/pill/label/VAIPE_P_{i}_{j}.json".format(i=i, j=j)
        if os.path.exists(json_path):
            pill_json = json.load(open(json_path))
            for pill in pill_json:
                pill_name_set.add(pill["label"])
    pill_all.append(list(pill_name_set))
            
            
```

```python
pill_all_string = []
for pill_line in pill_all:

    pill_line = [str(pill)+"_" for pill in pill_line]
    pill_all_string.append(" ".join(pill_line))
pill_all_string
```

```python
#import ifidf
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
tfidf.fit(pill_all_string)
pill_tfidf = tfidf.transform(pill_all_string)
pill_tfidf

```

```python
matrix_tfidf =  pill_tfidf.toarray()
print("matrix_tfidf.shape", matrix_tfidf.shape)

```

```python
matrix_tfidf_vocab = tfidf.vocabulary_
matrix_tfidf_vocab
```

```python
matrix_adj = np.zeros((108,108))
for pill_line, matrix_tfidf_line in zip(pill_all, matrix_tfidf):
    for i in range(len(pill_line) - 1):
        for j in range(i+1, len(pill_line)):
            matrix_adj[pill_line[i], pill_line[j]] = matrix_tfidf_line[matrix_tfidf_vocab[str(pill_line[i])+"_"]] + matrix_tfidf_line[matrix_tfidf_vocab[str(pill_line[j])+"_"]]
            matrix_adj[pill_line[j], pill_line[i]] = matrix_adj[pill_line[i], pill_line[j]]
        
        
```
